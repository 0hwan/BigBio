- 1) 정규분포 확률 밀도함수
    - mean = \\( \mu \\)
    - sd   = \\( \sigma \\)
       
$$ f(x) = \frac{1}{\sqrt{2\pi} \sigma } exp(- \frac{(x-\mu)^{2}}{2 \sigma^{2}})  $$

$$ = dnorm(x, mean = \mu, sd = \sigma) $$

```r
x = 0
mu   = 0
std  = 1

y = 1/(sqrt(2*pi)*std) *exp(-(x-mu)^2/2*std^2)
print (y)
```

## 문제!

통계학 성적 평균 50점, 표준편차 12점인 정규분포를 따른다고 한다. 
- 점수(확률변수)가 30점 이하일 확률?

```r
x = seq( 0, 100, .01 )
y = dnorm(x, mean=50, sd=12)
plot(x,y,type="l", main='f(x) = dnorm')

q = 30
polygon( c(0, x[x<q], q ), c(0, y[x<q] ,0 ), col="gray")
abline(v=q, h=0)
```

```r
q = 30
integrate( function(x) dnorm(x,mean=50,sd=12) , -Inf, q)
```

```r
pnorm(q, mean=50,sd=12)
```

```r
args(pnorm)
```

**분위수**(Quantil)?
   - 확률변수 X의 **p 분위수**는, $$ F(q) = P(X \leq q) = \int_{-\inf}^{q} f(x)\ dx = p $$ 를 만족하는 q 값

- pnorm(q) = p
- q = qnorm(p)

```r
p = .05
q = qnorm(p , mean=50, sd=12)

cat('Answer: ', q)
```

## 문제!

통계학 성적 평균 50점, 표준편차 12점인 정규분포를 따른다고 한다. 
- 상위 10%에 들기 위한 점수(확률변수)?

```r
p = 0.1
q = qnorm(p, mean=70, sd=12, lower.tail=F)

cat('Answer: ', q)
```

### 데이터 유형에 따라 시각화하는 방식이 다르다

- 1) 범주형 categorical data
    - 1-1) barplot
    - 1-2) pieplot
    - 1-3) dotplot
    
- 2) 수치형 numeric data
    - 2-1) histogram
    - 2-2) boxplot
    - 2-3) stripcharts

```r
d1 = table(juul3$sex)
print( d1 )
```

```r
d2 = with(juul3, table(sex, tanner))
print( d2 )
```

1-1) barplot

```r
par(mfrow=c(2,2))

# plot1
barplot(d1)
# plot2
barplot(d2)
# plot3
barplot(d2, legend.text=T)
# plot4
barplot(d2, beside=T)
```

```r
par(mfrow=c(1,2))
# plot1
pie(d1)

# plot2
dotchart(d2)
```

2-1) histogram

```r
d3 = juul$igf1
```

```r
par(mfrow=c(2,2))
hist(d3)
hist(d3, breaks=seq(0, 1000, 50))
hist(d3, breaks=5 )
hist(d3, freq=F)
```

2-2) boxplot

```r
par( mfrow=c(1,2) )

# plot1
boxplot(d3)

# plot2
group = juul2$sex
boxplot(d3~ group) # y~x ( y described using x )
```

2-3) stripchart

```r
par(mfrow=c(3,1))

# plot1
stripchart(d3~ group, main='plot1') 

# plot2
stripchart(d3~ group, method='stack' , main='plot2')

# plot3
stripchart(d3~ group, method='jitter', main='plot3')
```
